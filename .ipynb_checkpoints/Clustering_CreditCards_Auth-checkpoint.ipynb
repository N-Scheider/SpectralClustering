```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import umap.umap_ as umap
from sklearn.manifold import TSNE
import sklearn
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
from ProjetSpectralClustering import *
import warnings

warnings.filterwarnings('ignore')  # ignore notifications
```

```python
# Data set from https://www.kaggle.com/datasets/arjunbhasin2013/ccdata?resource=download
data = pd.read_csv("../SpectralClustering/CC_dataset.csv")
```

```python
del data["CUST_ID"]
# data.isna().sum() shows amount of NA in all variables
data.dropna(inplace=True) # removes observations with NA
data.reset_index(drop=True, inplace=True)
```

```python
data.head(5)
```

```python
# imported from https://www.kaggle.com/code/alisultanov/clustering-customer-personality-analysis/notebook
def scaling_func(df: pd.DataFrame) -> None:
    mms = MinMaxScaler() # from sklearn.prepocessing
    return pd.DataFrame(data=mms.fit_transform(df), columns=df.columns)


data_scaled = scaling_func(data)
data_scaled.index = data.index  # for convenient work with dataframes
```

```python
def dim_reduction(data: pd.DataFrame) -> pd.DataFrame:
    n_components = 8
    pca = PCA(n_components=n_components, random_state=42)
    data = pd.DataFrame(pca.fit_transform(data), columns =([f'PC{i}' for i in range(1, n_components + 1)]))
    return data

data_transform = dim_reduction(data_scaled)  # df: scaling + dimensionality reduction
data_transform.index = data_scaled.index  # for convenient work with dataframes
```

```python
sim_data = SimKNearestNeighborGraphs(data_scaled.to_numpy(dtype=float), 20)

k_grid = np.arange(2, 11)
dictionary_norm = list(np.repeat(0, len(k_grid)))
dictionary_normNG = list(np.repeat(0, len(k_grid)))
dictionary_unnorm = list(np.repeat(0, len(k_grid)))

for k in k_grid:
    dictionary_norm[k-2] = NormalizedSpectralClustering(sim_data, k)
    dictionary_normNG[k-2] = NormalizedNGSpectralClustering(sim_data, k)
    dictionary_unnorm[k-2] = UnormalizedSpectralClustering(sim_data, k)
```

```python
# Experimenting with TSNE library
# t-distributed stochastic neighbourhood embedding
# https://stats.stackexchange.com/questions/263539/clustering-on-the-output-of-t-sne gives good critical thinking about t-SNE
m = TSNE(learning_rate=50) # usually from 10 to 1000 where less means less adventuruous
tsne_features = m.fit_transform(data_scaled)
data_scaled["x"] = tsne_features[:,0]
data_scaled["y"] = tsne_features[:,1]
sns.scatterplot(x="x", y="y", hue='TENURE', data=data_scaled)
plt.show()
```

```python
ls = ["labelnorm", "labelnormng", "labelunnorm"]


for k in range(2,7):
    
    for i in ls:
        data_scaled[i] = 0

    for c in range(k):
        data_scaled["labelnorm"][dictionary_norm[k-2][c]] = c
        data_scaled["labelnormng"][dictionary_normNG[k-2][c]] = c
        data_scaled["labelunnorm"][dictionary_unnorm[k-2][c]] = c
    
    fig, axs = plt.subplots(1, 3, figsize=(10,4))
    fig.suptitle(f'Different Algos for {k} clusters')
    plt.subplots_adjust(wspace = 0.4)
    sns.scatterplot(x="x", y="y", hue='labelnorm', data=data_scaled, ax=axs[0], palette="Set1")
    sns.scatterplot(x="x", y="y", hue='labelnormng', data=data_scaled, ax=axs[1], palette="Set1")
    sns.scatterplot(x="x", y="y", hue='labelunnorm', data=data_scaled, ax=axs[2], palette="Set1")
    plt.show()

```

```python
# Experimenting with umap library
# uniform manifold approximation projection
# https://pair-code.github.io/understanding-umap/ good explanation and comparision to t-SNE
```

```python
mod = umap.UMAP(n_neighbors=250,min_dist=0.35,n_components=2,random_state=42)
embedding = pd.DataFrame(mod.fit_transform(data_scaled), columns = ['UMAP1','UMAP2'])
```

```python
data_scaled["umap1"] = embedding["UMAP1"]
data_scaled["umap2"] = embedding["UMAP2"]
# Produce sns.scatterplot and pass metadata.subclasses as color
sns_plot = sns.scatterplot(x="umap1", y="umap2", hue="TENURE", data=data_scaled, alpha=.1, linewidth=0, s=1)
sns_plot.legend(loc='center left', bbox_to_anchor=(1, .5))
plt.show()
```
